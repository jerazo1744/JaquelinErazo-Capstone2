Split the date into year, month, and day

```python
reviews[['review_year', 'review_month', "review_day"]]= reviews['date'].str.split('-', expand = True)
```

```python
reviews.head()
```

Filter the review date by the year 2019, rename columns, specify which columns I want returned as a df

```python
reviews2019 = reviews.loc[reviews.review_year == '2019'].reset_index()
reviews2019.rename(columns={'listing_id':'id'}, inplace=True)
reviews2019 = reviews2019[['id','review_year','review_month','review_day']]
reviews2019.head()
```

```python
listings_ext = listings_ext[['id', 'review_scores_rating']]
```

```python
listings_ext.head()
```

```python
listings_ext.info()
```

Join two df's: reviews2019 and listings_ext with reviews2019 being the left df/table ; name new df as review_ratings

```python
review_ratings = reviews2019.join(listings_ext.set_index('id'), on='id')
#review_ratings.info()
review_ratings
```

```python
review_ratings.loc[review_ratings.review_scores_rating.isnull()]
```

```python
listings_filter = listings[['id', 'neighbourhood', 'latitude', 'longitude', 'price', 'room_type']]
```

Join two df's: review_ratings and listings_filter with review_ratings as the left df/table ; new table is review_ratings_listing

```python
review_ratings_listing = review_ratings.join(listings_filter.set_index('id'), on='id')
```

Group by id and count how many times it occurs to find the total number of reviews in 2019, order it by descending and find the top 100 listings with the most reviews and reset the index

```python
top_100 = review_ratings_listing.groupby(['id']).count().sort_values(by='review_year', ascending=False).reset_index()
```

```python
top_100 = top_100.head(100)
top_100 = top_100[['id','review_year']]
top_100.rename(columns={'review_year':'2019_total_reviews'}, inplace=True)
top_100.head(1)
```

Add a column by joining on id from the listings_ext df for review ratings

```python
top_100 = top_100.join(listings_ext.set_index('id'), on='id')
top_100.head(1)
```

```python
listings = listings[['id', 'neighbourhood', 'latitude', 'longitude', 'room_type', 'price']]
listings.head(1)
```

Add the columns by joining on id for neighbourgood, lat, long, room type and price from listings df

```python
top_100 = top_100.join(listings.set_index('id'), on='id')
```

```python
top_100 = top_100[['id', 'neighbourhood', 'latitude', 'longitude', 'room_type', 'price', '2019_total_reviews', 'review_scores_rating']]
top_100
```
